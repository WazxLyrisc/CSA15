```python
import pandas as pd
```

```python
df = pd.read_csv("car-sales-data.csv", index_col=[0])
df
```

```python
df.dropna()
df.Colour = df.Colour.fillna(0)
df.Odometer = df.Odometer.fillna(0)
df.Doors = df.Doors.fillna(0)
df.Price = df.Price.fillna(0)
df.tail()
```

```python
# df = pd.read_csv("car-sales-data.csv", index_col=[0])
# df

# 1. Chiếc xe nào được bán với giá đắt nhất?
# 2. Hãng xe nào có chiếc xe với giá trên mỗi kilomet đã đi được rẻ nhất?
# 3. Trung bình giá xe của các hãng xe là bao nhiêu? Hãng xe nào có trung bình giá bán cao nhất?

# df['Price'] = df['Price'].apply(lambda x : x[1: -1])
df['Price'] = pd.to_numeric(df['Price'].replace('[\$,]', '', regex=True), errors='coerce')
# df['Price'] = pd.to_numeric(df['Price'])
# df.dtypes()

most_expensive_car = df.loc[df['Price'].idxmax()]
most_expensive_car
```

```python
# 2. Hãng xe nào có chiếc xe với giá trên mỗi kilomet đã đi được rẻ nhất?
df['Price_per_km'] = df['Price'] / df['Odometer']
cheapest_per_km = df.loc[df['Price_per_km'].idxmin()]
cheapest_per_km
```

```python
# 3. Trung bình giá xe của các hãng xe là bao nhiêu? Hãng xe nào có trung bình giá bán cao nhất?
df_price = df.groupby("Make")["Price"].mean().sort_values(ascending=False)
print("Trung bình giá xe theo hãng:\n", df_price)
print("Hãng có giá trung bình cao nhất:", df_price.idxmax())
```

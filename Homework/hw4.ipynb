```python
# Có bao nhiêu người trên 30 tuổi
df = pd.read_csv("hw3_data.csv")

age_data = df[["Name", "Age"]]
age_data[df["Age"] >= 30]
```

```python
# Tổng số tiền lương của tất cả mọi người có trong dữ liệu là bao nhiêu
sum_salary = df["Salary"].sum()
print("Tổng số tiền lương của tất cả mọi người là: ", sum_salary)
```

```python
# Khởi tạo series từ list
fruits = pd.Series(["apple", "orange", "banana", "Cherry"])
fruits
```

```python
# Thay đổi chỉ số
fruits = pd.Series(["apple", "orange", "banana", "Cherry"], index=["A", "B", "C", "D"])
fruits
```

```python
# KHởi tạo Series từ dictionary
fruit_dict = {
    'A': 'apple',
    'B': 'banana',
    'C': 'cherry',
    'D': 'orange'
}

fruits = pd.Series(fruit_dict)
fruits
```

```python
temperature_dict = {
    "Hà Nội": 28,
    "Đà Nẵng": 30,
    "TP. Hồ Chí Minh": 35,
    "Huế": 29,
    "Hải Phòng": 26,
    "Nghệ An": 30,
    "Lâm Đồng": 39
}
temperature = pd.Series(temperature_dict)
temperature
```

```python
# Một số hàm thống kê nhanh
print(temperature.max()) #Lấy giá trị lớn nhất
print(temperature.min()) #Lấy giá trị bé nhất
print(temperature.sum()) #Lấy tổng
print(temperature.mean()) #Lấy trung bình cộng
print(temperature.median()) #Lấy trung vị
print(temperature.mode()) #Lấy giá trị xuất hiện nhiều nhất
```

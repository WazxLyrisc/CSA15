```python
# Khởi tạo data bằng dictionary
animals = pd.DataFrame({
    "dong_vat" : ["Sao La", "Khỉ đuôi ngắn", "Voi châu Á"],
    "tuoi_tho" : [25,25,70],
    "loai" : ["có vú", "linh trưởng", "có vú"],
    "thuc_an" : ["thực vật", "thực vật", "thực vật"],
    "moi_truong_song" : ["rừng", "rừng", "thảo nguyên"],
    "tap_tinh" : ["Bầy đàn", "Bầy đàn", "Bầy đàn"]
})

animals
```

```python
# Hiển thị thông tin về bảng (số cột, số hàng, kiểu dữ liệu)
animals.info()
```

```python
import pandas as pd
df = pd.read_csv('spotify.csv')
df
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d")
df.dtypes
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,7))
df_line = df.set_index("Date")
plt.plot(df_line) # Váº½ biá»ƒu Ä‘á»“ Ä‘Æ°á»ng cho toÃ n bá»™ cÃ¡c cá»™t trong biá»ƒu Ä‘á»“
plt.legend(df_line.columns) # HIá»ƒn thá»‹ chÃº thÃ­ch cho tá»«ng Ä‘Æ°á»ng
plt.ylabel("Views")
plt.title("Spotify song views from January 2017 to January 2018") #GÃ¡n tÃªn cho biá»ƒu Ä‘á»“
plt.show() #Hiá»ƒn thá»‹ biá»ƒu Ä‘á»“
```

```python
df = pd.read_csv("2024 Billionaire List.csv", index_col=[0])
df["2024 Net Worth"] = df["2024 Net Worth"].apply(lambda x : x[1 : -1])
df["2024 Net Worth"] = pd.to_numeric(df["2024 Net Worth"])
df_industry = df[["Industry", "2024 Net Worth"]].groupby("Industry").sum()
df_industry = df_industry.sort_values(by="2024 Net Worth", ascending=False)
# df_industry.head(10)
df_top_industry = df_industry[:10]

# Táº¡o biá»ƒu Ä‘á»“ cá»™t vá»›i trá»¥c ngang lÃ  industry, trá»¥c dá»c vÃ  net Worth
plt.bar(x=df_top_industry.index, height=df_top_industry["2024 Net Worth"])
plt.ylabel("Net Worth ($B)")
plt.title("Top 10 Net Worth by Industry")
plt.show()
```

```python
df["Age"] = df["Age"].astype('Int64') 
plt.hist(
    x = df["Age"],
    edgecolor = 'white'
)

plt.title("Forbes's 2024 Billionaires Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
# Xu hÆ°á»›ng phÃ¢n bá»‘ Ä‘á»™ tuá»•i cá»§a cÃ¡c tá»· phÃº trong khoáº£ng 60 - 80 tuá»•i lÃ  nhiá»u nháº¥t
```

```python
plt.scatter(
    x=df["Age"],
    y = df["2024 Net Worth"],
    s = 30,
    # s = df["2024 Net Worth"] * 10,
    linewidths = 0.1, #Äá»˜ má»
    edgecolors = 'black'
)

plt.title("2024 Billionares's Age w.r.t Net Worth")
plt.xlabel("Age")
plt.ylabel("Net Worth ($B)")
plt.show()
```

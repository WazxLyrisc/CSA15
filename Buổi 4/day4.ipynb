```python
import pandas as pd
```

```python
df = pd.read_csv("2024 Billionaire List.csv", index_col=[0])
df["2024 Net Worth"] = df["2024 Net Worth"].apply(lambda x : x[1 : -1])
df["2024 Net Worth"] = pd.to_numeric(df["2024 Net Worth"])
df.dtypes
# df.head()
```

```python
df["Age"] = df["Age"].astype('Int64') #Chuyển dữ liệu cột từ float64 -> int64
df.dtypes
# df.head()
```

```python
# Lấy ra 2 cột industry và 2024 net worth => nhóm theo cột industry
df[["Industry", "2024 Net Worth"]].groupby("Industry")
```

```python
# Tính tổng giá trị từng industry
df[["Industry", "2024 Net Worth"]].groupby("Industry").sum()
```

```python
# Sắp xếp cột 2024 Net Worth theo thứ tự giảm dần
df_industry = df[["Industry", "2024 Net Worth"]].groupby("Industry").sum()
df_industry = df_industry.sort_values(by="2024 Net Worth", ascending=False)
df_industry
```

```python
import pycountry_convert as pc
def get_continent(country_name):
    try:
        country_code = pc.country_name_to_country_alpha2(country_name, cn_name_format="default")
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except:
        return "Unknown"
    
df["Continent"] = df["Citizenship"].apply(get_continent)
df[["Continent", "2024 Net Worth"]].groupby("Continent").mean()
```

```python
df.dropna() #Bỏ tất cả các hàng có giá trị bị khuyết thiếu
df.dropna(axis=1) #Loại bỏ tất cả các cột
df.isnull().sum() # Đếm số giá trị NaN

df.drop(["birthdate ", "birthplace", "weight"], axis=1, inplace=True) # Xoá cột
df

df[df.goals.isna()] # Lấy ra giá trị NaN bị lỗi

df.goals = df.goals.fillna(0) #THay thế NaN thành số 0
df.tail()

df.duplicated().sum() # đếm số hàng trùng
df = df.drop_duplicates() # Xử lý hàng trùng
```
